```python
tasks = []
```

```python
task1 = input("Enter first task: ")
tasks.append(task1)

task2 = input("Enter second task: ")
tasks.append(task2)

print("Task added successfully")
```

```python
print("My To-Do List:")

for i, task in enumerate(tasks, start=1):
    print(i, ".",task)
```

```python
num = int(input("Enter task number to delete: "))
tasks.pop(num -1)
print("Task deleted successfully")
```

```python
print("My To-Do List:")

for i, task in enumerate(tasks, start=1):
    print(i, ".",task)
```
