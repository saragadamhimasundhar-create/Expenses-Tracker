```python
import csv
from datetime import datetime
```

```python
expenses = []
```

```python
date1 = datetime.now().strftime("%d-%m-%y")
category1 = input("Enter category(Food/Travel/Shopping/etc):")

name1 = input("Enter 1st expense name: ")
amount1 = float(input("Enter amount1: "))
expenses.append([date1,category1,name1, amount1])

date2 = datetime.now().strftime("%d-%m-%y")
category2 = input("Enter category(Food/Travel/Shopping/etc):")

name2 = input("Enter 2nd expense name: ")
amount2 = float(input("Enter amount2: "))
expenses.append([date2,category2,name2, amount2])

print("Expense added successfully!")
```

```python
print("Expense List")
print("-"*50)

for i, expense in enumerate(expenses, start=1):
    print(i, ".", expense[0], "|", expense[1],"|",expense[2],"|rs", expense[3])
```

```python
total = 0

for expense in expenses:
    total += expense[3]

print("Total Expense = rs", total)
```

```python
search = input("Enter category to search: ")

print("\nMatching Expenses:")

for expense in expenses:
    if expense[1].lower() == search.lower():
        
        print(expense[0], "|", expense[1], "|", expense[2], "| rs", expense[3])
```

```python
num = int(input("Enter expense number to delete: "))

expenses.pop(num - 1)

print("Expense deleted successfully!")
```

```python
print("Expense List")

for i, expense in enumerate(expenses, start=1):
    print(i, ".", expense[0], "- ₹", expense[1])
```

```python
total = 0

for expense in expenses:
    total += expense[3]

print("Total Expense = ₹", total)
```
